```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize = (9, 3))

# subplot1
plt.subplot(131)
plt.bar(names, values)

# subplot2
plt.subplot(132)
plt.scatter(names, values)

# subplot3
plt.subplot(133)
plt.plot(names, values)

plt.suptitle("Categorical Plotting")
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)
```

```python
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2 * np.pi * t2), 'r--')
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1, 2, 3])
plt.subplot(212)
plt.plot([4, 5, 6])

plt.figure(2)
plt.plot([4, 5, 6])

plt.figure(1)

plt.subplot(211)

plt.title('Easy as 1, 2, 3')
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
x
```

```python
n, bins, patches = plt.hist(x, 50, density = True, facecolor = 'g', alpha = 0.75)

# plt.xlabel('Smarts')
plt.xlabel('My Data (Smarts)', fontsize = 14, color = 'orange')
# plt.ylabel('Probability')
plt.ylabel('Probability', fontsize = 14,  color = 'orange')
# plt.title('Histogram of IQ', fontsize = 14, color = 'green')
plt.title(r'$E=mc^2$')
plt.text(60, .025, r'$\mu = 100,\ \sigma = 15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line, = plt.plot(t, s, lw = 2)

plt.annotate('local max', xy = (2, 1), xytext = (3, 1.5),
             arrowprops = dict(facecolor = 'black',shrink = 0.03),
            )

plt.ylim(-2, 2)
plt.show()
```

```python
plt.xscale('log')

np.random.seed(19680801)

y = np.random.normal(loc = 0.5, scale = 0.4, size = 1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

plt.figure()

plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title("log")
plt.grid(True)

plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh = 0.01)
plt.title('symlog')
plt.grid(True)

plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)

plt.subplots_adjust(top = 0.92, bottom = 0.08, left = 0.10, right = 0.95, hspace = 0.25, wspace = 0.35)
plt.show()
```
