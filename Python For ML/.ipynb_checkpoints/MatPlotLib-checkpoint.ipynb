# MatPlotLib

> Source: https://matplotlib.org/stable/tutorials/index.html

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel("Prices")
plt.show()
```

```python
plt.plot([1, 3, 4, 9], [-1, 8, 1, 2], 'ro')
```

```python
plt.plot([1, 3, 4, 9], [-1, 8, 1, 2], 'b.-.')
```

```python
plt.plot([1, 3, 4, 9], [-1, 8, 1, 2], 'g^')
```

```python
t = np.arange(0, 5, 0.2)
plt.plot(t, t, "r-.",t,  t**2, 'bs', t, t**3, "g^")
```

```python
data = {
    'a' : np.arange(500),
    'b' : np.random.randint(0, 50, 50),
    'c' : np.random.randn(500)
}
data
```

```python
data['a']   # 'a' : np.arange(500)
```

```python
data['b']   # 'b' : np.random.randint(0, 50, 50)
```

```python
data['c']   # 'c' : np.random.randn(500)
```

```python
plt.scatter(data["a"], data["c"])
plt.xlabel("A")  # Label for x-axis
plt.ylabel("C")  # Label for y-axis
plt.show()
```
